# Accessing Euclid Q1 data in the cloud

**AAS workshop:** *Euclid Data in the Cloud - Access, Analysis, and Science Opportunities* (90 min)

_Notebook 1 of 3._ This is the teaching notebook for the data-access half of the workshop. The aim is to walk through, from scratch, how to pull each Euclid Q1 data product out of IRSA without downloading the whole release:

- **TAP** queries against IRSA's catalog service (MER photometry, SPE line features, PHZ redshifts, ...)
- **SIA + IBE** server-side image cutouts from MER mosaics
- **Direct anonymous S3** reads of SIR 1D spectra via `fsspec`

Notebooks 2 and 3 (`02_Euclid_clusters.ipynb`, `03_Euclid_AGN_assignment.ipynb`) put these tools to work on science problems.
